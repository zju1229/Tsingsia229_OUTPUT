### 3. 傅里叶系数计算 （50分）

#### 基础知识
由傅里叶级数知识我们可以知道，在$[-\pi,\pi]$上的函数$f(x)$可以定义如下的傅里叶级数：
$$
S(f)(x) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left( a_n \cos(nx) + b_n \sin(nx) \right), S_N(f)(x) = \frac{a_0}{2} + \sum_{n=1}^{N} \left( a_n \cos(nx) + b_n \sin(nx) \right).
$$
其中各项系数的计算公式为：
$$
a_0 = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \mathrm{d}x, \quad a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \cos(nx) \mathrm{d}x, \quad b_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \sin(nx) \mathrm{d}x.
$$

$S_N(f)$ 是否收敛于$f(x)$ 这个问题比较复杂，这里不赘述，但是有几个结论我们可以给出：
1. 如果$f(x) \in C^2$, 则$S_N(f)(x)$一致收敛于$f(x)$
2. 存在连续的$f(x)$，它的傅里叶级数不是逐点收敛于$f(x)$，进一步分析可以证明，满足这种性质的函数在$C([-\pi,\pi])$中是稠密的。

以上 $C^p(\Omega)$ 表示在$\Omega$上具有$p$阶连续导数的函数，$C(\Omega)$表示在$\Omega$上连续的函数，并定义 $C^{-1}(\Omega)$ 为在 $\Omega$ 上不连续的函数。

定义以下评价函数(以下积分可用等距节点近似计算)：
$$
d(f) = \left(\int_{-\pi}^{\pi} |f(x)|^2 \mathrm{d}x \right)^{\frac{1}{2}}. 
$$

#### 任务
1. 计算函数
$$
\begin{cases}
f(x) = -1, & x \in [-\pi,0) \\ f(x) = 1, & x \in [0,\pi]
\end{cases}
$$
在$[-\pi,\pi]$上的傅里叶级数展开式，计算 $a_n, b_n$ 的值，并分别画出前 $n=5,10,20$ 项的傅里叶级数图像（$f(x)$ 也要画，可以画在一张图里），并计算 $d(f-S_5(f))$，$d(f-S_{10}(f))$，$d(f-S_{20}(f))$ 的值。

2. 计算函数
$$
f(x) = \frac{|x|}{\pi}, \quad x \in [-\pi,\pi]
$$
在$[-\pi,\pi]$上的傅里叶级数展开式，计算 $a_n, b_n$ 的值，并分别画出前 $n=5,10,20$ 项的傅里叶级数图像（$f(x)$ 也要画，可以画在一张图里），并计算 $d(f-S_5(f))$，$d(f-S_{10}(f))$，$d(f-S_{20}(f))$ 的值。

3. 计算函数
$$
f(x) = \frac{x^2 |x|}{\pi^3}, \quad x \in [-\pi,\pi]
$$
在$[-\pi,\pi]$上的傅里叶级数展开式，计算 $a_n, b_n$ 的值，并分别画出前 $n=5,10,20$ 项的傅里叶级数图像（$f(x)$ 也要画，可以画在一张图里），并计算 $d(f-S_5(f))$，$d(f-S_{10}(f))$，$d(f-S_{20}(f))$ 的值。

```python
# 在这里作答第三问
# ================== 导入库 ================== 
from IPython.display import display, Math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
from sympy import (
    init_printing, symbols, Piecewise, pi, integrate,
    sin, cos, sqrt, simplify, Abs, lambdify
)
init_printing(use_latex=True)  # 启用 LaTeX 渲染

# ================== 符号定义 ================== 
x = symbols('x')
n = symbols('n',integer=True,nonnegative=True)

# ================== 函数定义 ==================
def fourier_an(f):
    a_n = (1/pi) * integrate(f * cos(n*x), (x, -pi, pi)).simplify()
    b_n = (1/pi) * integrate(f * sin(n*x), (x, -pi, pi)).simplify()
    display(Math(r'a_n ='))
    display(a_n)
    display(Math(r'b_n ='))
    display(b_n)
def fourier_sum(c, f, N):
    """计算傅里叶级数"""
    a0 = (1/pi) * integrate(f, (x, -pi, pi)).simplify()
    s = a0 / 2
    for i in range(1, N+1):
        a, b = 0, 0
        if c != 1:  # 跳过余弦项的条件
            a = (1/pi) * integrate(f * cos(i*x), (x, -pi, pi)).simplify()
        if c != 2:  # 跳过正弦项的条件
            b = (1/pi) * integrate(f * sin(i*x), (x, -pi, pi)).simplify()
        s += a * cos(i*x) + b * sin(i*x)
    return simplify(s)

def d(func):
    """计算 L2 误差范数"""
    numfunc = lambdify(x, Abs(func)**2, 'numpy')
    integral_value, _ = quad(numfunc, -np.pi, np.pi)
    return np.sqrt(integral_value).round(5)

def funcdraw(f, c, Ns, title):
    """绘制函数与傅里叶级数对比图"""
    xx = np.linspace(-np.pi, np.pi, 1000)
    fig, ax = plt.subplots(figsize=(10, 6))
    f_numeric = lambdify(x, f, 'numpy')
    ax.plot(xx, f_numeric(xx), label='f(x)', lw=2, color='black')
    for N in Ns:
        S_N = fourier_sum(c, f, N)
        S_N_numeric = lambdify(x, S_N, 'numpy')
        ax.plot(xx, S_N_numeric(xx), label=f'N={N}', ls='--', alpha=0.7)
    ax.legend()
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    plt.show()

# ================== 任务1：方波函数 ================== 
display(Math(r'====== Task 1 ======'))
f = Piecewise((-1, x < 0), (1, x >= 0))
fourier_an(f)
S1_5 = fourier_sum(1, f, 5)
S1_10 = fourier_sum(1, f, 10)
S1_20 = fourier_sum(1, f, 20)

display(Math(r'S_5(x):'), S1_5)
display(Math(r'S_{10}(x):'), S1_10)
display(Math(r'S_{20}(x):'), S1_20)

funcdraw(f, c=1, Ns=[5, 10, 20], title='Task 1: Fourier_sum Picture')
display(Math(r'd_1 ='), d(f - S1_5))
display(Math(r'd_2 ='), d(f - S1_10))
display(Math(r'd_3 ='), d(f - S1_20))

# ================== 任务2：绝对值函数 ================== 
display(Math(r'====== Task 2 ======'))
g = Abs(x) / pi
fourier_an(g)
S2_5 = fourier_sum(2, g, 5)
S2_10 = fourier_sum(2, g, 10)
S2_20 = fourier_sum(2, g, 20)

display(Math(r'S_5(x):'), S2_5)
display(Math(r'S_{10}(x):'), S2_10)
display(Math(r'S_{20}(x):'), S2_20)

funcdraw(g, c=2, Ns=[5, 10, 20], title='Task 2: Fourier_sum Picture')
display(Math(r'd_1 ='), d(g - S2_5))
display(Math(r'd_2 ='), d(g - S2_10))
display(Math(r'd_3 ='), d(g - S2_20))

# ================== 任务3：高阶奇函数 ================== 
display(Math(r'====== Task 3 ======'))
h = (x**2) * Abs(x) / (pi**3)
fourier_an(h)
S3_5 = fourier_sum(2, h, 5)
S3_10 = fourier_sum(2, h, 10)
S3_20 = fourier_sum(2, h, 20)

display(Math(r'S_5(x):'), S3_5)
display(Math(r'S_{10}(x):'), S3_10)
display(Math(r'S_{20}(x):'), S3_20)

funcdraw(h, c=2, Ns=[5, 10, 20], title='Task 3: Fourier_sum Picture')
display(Math(r'd_1 ='), d(h - S3_5))
display(Math(r'd_2 ='), d(h - S3_10))
display(Math(r'd_3 ='), d(h - S3_20))
```

### 4. 进一步探讨 （20分）
在考虑函数 $f(x)$ 在给定区间上的傅里叶级数 $\sum a_n \cos(nx) + b_n \sin(nx)$ 的收敛时，$f(x)$ 的光滑性($f\in C^p(\mathbb{R})$，$p$的大小)与级数的收敛速度是否有一定的关系？给出一个你自己的理解和描述。

### 

#### 第四问
### 一、理论依据
1. **光滑性**的定义：
  - 若 $f \in {C}^{p}(\mathbb{R})$ ，表示 $f(x)$ 在实数域上具有 $p$ 阶连续导数。
  - 光滑性越高( $p$ 越大),函数的高阶导数越平滑，无突变或不连续点。
2. **傅里叶系数的衰减速率**：
   - 对于 $$ f \in C^p $$ ，其傅里叶系数 $a_n$ 和 $b_n$ 的衰减速率与 $p$ 直接相关：  
   $$
   |a_n|, |b_n| = O\left(\frac{1}{n^{p+1}}\right).
   $$
   - **推导**：通过 $ p $ 次分部积分，可将傅里叶系数与 $ f^{(p)}(x) $ 的积分关联。由于 $ f^{(p)}(x) $ 连续，其积分值为有界量，从而系数衰减速率由 $ n^{-(p+1)} $ 主导。

3. **收敛性分类**：  
   - **低光滑性（ $ p=0 $ )**：系数衰减为 $ O(1/n) $ ，级数收敛缓慢，可能出现吉布斯现象。  
   - **中等光滑性（ $ p \geq 1 $ )**：系数衰减为 $ O(1/n^{p+1}) $ ，收敛速度显著提升。  
   - **高光滑性（ $ p \to \infty $ )**：系数以指数速率衰减（如 $ O(e^{-kn}) $ ），级数几乎立即收敛。

---

#### **2. 实例验证**

1. **分段常数函数（$ p=0 $)**：  
   - 函数 $ f(x) = \text{sgn}(x) $（符号函数）在 $ x=0 $ 处不连续，其傅里叶系数 $ b_n \sim O(1/n) $。  
   - **结果**：收敛缓慢，且在 $ x=0 $ 附近出现吉布斯现象（超调量约8.95%）。

2. **三角波函数（$ p=1 $)**：  
   - 函数 $ f(x) = |x| $ 连续但一阶导数不连续，其傅里叶系数 $ a_n \sim O(1/n^2) $。  
   - **结果**：收敛速度优于分段函数，但仍存在轻微振荡。

3. **高斯函数（$ p=\infty $)**：  
   - 函数 $ f(x) = e^{-x^2} $ 无限光滑，其傅里叶系数以指数速率衰减。  
   - **结果**：级数迅速收敛，无吉布斯现象。

---

#### **3. 数学推导**

以 $ f \in C^1(\mathbb{R}) $ 为例，通过分部积分分析系数衰减：  
$$
a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \cos(nx) \, dx.
$$
令 $ u = f(x) $，$ dv = \cos(nx) dx $，则：
$$
a_n = \frac{1}{n\pi} \left[ f(x) \sin(nx) \bigg|_{-\pi}^{\pi} - \int_{-\pi}^{\pi} f'(x) \sin(nx) \, dx \right].
$$
由于 $ f \in C^1 $，边界项 $ f(x) \sin(nx) \bigg|_{-\pi}^{\pi} = 0 $，故：
$$
a_n = -\frac{1}{n\pi} \int_{-\pi}^{\pi} f'(x) \sin(nx) \, dx.
$$
若 $ f' \in C^1 $，可继续分部积分，得到：
$$
a_n = O\left(\frac{1}{n^2}\right).
$$

---

#### **总结**

函数的光滑性 $p$ 与傅里叶级数的收敛速度成严格正相关：  

$$
\text{收敛速度} \propto \frac{1}{n^{p+1}}.
$$
